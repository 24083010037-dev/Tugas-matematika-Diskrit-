###1. Tabel pivot konsonan dan vokal dalam kata

```python
words = re.findall(r"[a-zA-Z]+", text.lower())

def count_vowels(w): return sum(1 for ch in w if ch in "aeiou")
def count_consonants(w): return sum(1 for ch in w if ch in "bcdfghjklmnpqrstvwxyz")

df = pd.DataFrame({
    "word": words,
    "vowels": [count_vowels(w) for w in words],
    "consonants": [count_consonants(w) for w in words]
})

pivot = pd.crosstab(df["vowels"], df["consonants"])

#1. Heatmap
plt.figure(figsize=(10,5))
sns.heatmap(pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Heatmap Kombinasi Jumlah Vokal–Konsonan per Kata")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()
```

```python
# 2. 3D Surface Plot
X, Y = np.meshgrid(pivot.columns, pivot.index)
Z = pivot.values

fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap="plasma", edgecolor='k', alpha=0.9)
ax.set_xlabel("Jumlah Konsonan")
ax.set_ylabel("Jumlah Vokal")
ax.set_zlabel("Frekuensi")
ax.set_title("Distribusi Frekuensi Kata (3D Surface)")
fig.colorbar(surf, shrink=0.6, aspect=10, pad=0.1)
plt.show()
```

##Surface plot 3D
Pada distribusi Frekuensi Kata 3D ini menunjukan puncak tertinggi berwana (kuning) jenis kata dengan struktur tersebut Sumbu X menunjukkan jumlah huruf konsonan dalam setiap kata, sumbu Y menunjukkan jumlah huruf vokal, dan sumbu Z (tinggi permukaan) menunjukkan frekuensi kemunculan kata dengan kombinasi tersebut. Permukaan grafik berwarna dari ungu hingga kuning menggunakan skema warna plasma di mana warna kuning menandakan frekuensi tertinggi, sedangkan ungu menunjukkan frekuensi terendah.

Grafik 3D memperlihatkan permukaan dengan puncak (warna kuning) di area:

- 1 vokal dan 2 konsonan, dengan frekuensi sekitar 313 kata (tertinggi).

- 1 vokal dan 3 konsonan, dengan frekuensi sekitar 234 kata.

Sementara itu, bagian bawah grafik (berwarna ungu gelap) menunjukkan, yaitu kata yang memiliki lebih dari 6 konsonan atau lebih dari 4 vokal.

```python
import re
sentences = re.split(r'[.!?]+', text)
sentence_lengths = [len(re.findall(r"[a-zA-Z]+", s)) for s in sentences if s.strip()]
df_sent = pd.DataFrame({"sentence_index": range(1, len(sentence_lengths)+1),
                        "sentence_length": sentence_lengths})
plt.figure(figsize=(8,5))
sns.histplot(df_sent["sentence_length"], bins=10, kde=True, color="skyblue", edgecolor="black")
plt.title("Histogram Jumlah Kata per Kalimat")
plt.xlabel("Jumlah Kata per Kalimat")
plt.ylabel("Frekuensi")
plt.show()
```

Kurva biru halus di atas batang menggambarkan pola distribusi data.

Berdasarkan hasil histogram, terlihat bahwa distribusi jumlah kata per kalimat tidak merata dan cenderung tidak seragam. Terdapat dua kelompok utama yang menonjol, yakni kalimat dengan sekitar 400 kata dan sekitar 1300 kata. Hal ini menunjukkan bahwa teks yang dianalisis terdiri dari dua jenis kalimat dengan panjang yang sangat berbeda beberapa kalimat sangat pendek, sementara yang lain jauh lebih panjang. Rentang di antara kedua nilai tersebut (400–1200 kata) tampak hampir kosong, menandakan tidak ada kalimat dengan panjang menengah.

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df_sent["sentence_length"], color="lightblue", width=0.3)
sns.swarmplot(y=df_sent["sentence_length"], color="darkblue", alpha=0.7)
plt.title("Boxplot + Swarmplot: Distribusi Jumlah Kata per Kalimat")
plt.ylabel("Jumlah Kata per Kalimat")
plt.show()
```

##Boxplot + Swarmplot
Gambar di atas merupakan gabungan dari boxplot dan swarmplot. Visualisasi ini menampilkan distribusi jumlah kata per kalimat yang bertujuan untuk melihat variasi panjang kalimat di seluruh teks.

Seperti terlihat pada gambar, sebagian besar kalimat memiliki panjang antara sekitar 600 hingga 1100 kata, dengan median berada di sekitar 850 kata per kalimat. Terdapat juga dua titik ekstrem (outlier), yaitu satu kalimat dengan panjang sekitar 400 kata dan satu lagi mencapai sekitar 1300 kata.

Rentang interkuartil (IQR) menunjukkan bahwa 50% kalimat memiliki panjang antara kurang lebih 650 hingga 1100 kata, menandakan variasi yang cukup besar antar kalimat.

###Jumlah kalimat dalam paragraf

```python
paragraphs = [p.strip() for p in text.split("\n") if p.strip()]
sentences_per_paragraph = [len(re.split(r'[.!?]+', p)) - 1 for p in paragraphs]
df_para = pd.DataFrame({"paragraph_index": range(1, len(sentences_per_paragraph)+1),
                        "sentence_count": sentences_per_paragraph})

matrix = np.zeros((len(df_para), len(df_para)))
for i in range(len(df_para)):
    for j in range(len(df_para)):
        matrix[i, j] = abs(df_para["sentence_count"][i] - df_para["sentence_count"][j])

plt.figure(figsize=(8,6))
sns.heatmap(matrix, annot=True, fmt=".0f", cmap="YlGnBu",
            xticklabels=df_para["paragraph_index"],
            yticklabels=df_para["paragraph_index"])
plt.title("Heatmap Variasi Jumlah Kalimat Antar Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Paragraf ke-")
plt.show()
```

##Headmap
Heatmap di atas menampilkan variasi jumlah kalimat antar paragraf dalam sebuah teks. Setiap kotak pada heatmap menunjukkan selisih jumlah kalimat antara dua paragraf yang dibandingkan. Angka 0 pada diagonal utama menunjukkan bahwa paragraf dibandingkan dengan dirinya sendiri, sehingga perbedaannya tidak ada atau nol. Semakin besar nilai yang muncul, semakin besar pula perbedaan jumlah kalimat antar paragraf tersebut. Warna pada heatmap menggambarkan tingkat variasinya.

- Ukuran matriks 20×20 menunjukkan bahwa teks terdiri dari 20 paragraf.

- Nilai selisih antar paragraf berkisar antara 0 hingga 1, yang berarti panjang kalimat antar paragraf relatif mirip.

- Warna biru tua (nilai 1) menunjukkan pasangan paragraf dengan perbedaan jumlah kalimat terbesar.

- Warna kuning muda (nilai 0) mendominasi heatmap, menandakan bahwa sebagian besar paragraf memiliki jumlah kalimat yang hampir sama atau seimbang.

```python
plt.figure(figsize=(9,6))

sizes = df_para["sentence_count"] * 50
colors = df_para["sentence_count"]

plt.scatter(df_para["paragraph_index"], df_para["sentence_count"],
            s=sizes, c=colors, cmap="coolwarm", alpha=0.7, edgecolors="black")

plt.title("Bubble Chart: Panjang Paragraf vs Posisi Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Jumlah Kalimat")
plt.colorbar(label="Jumlah Kalimat")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

Sumbu X menunjukkan urutan paragraf (paragraf sekitar ke 30 ), sedangkan sumbu Y menunjukkan jumlah kalimat di setiap paragraf.
