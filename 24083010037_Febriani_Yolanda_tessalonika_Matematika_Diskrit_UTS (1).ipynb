###1. Tabel pivot konsonan dan vokal dalam kata

```python
words = re.findall(r"[a-zA-Z]+", text.lower())

def count_vowels(w): return sum(1 for ch in w if ch in "aeiou")
def count_consonants(w): return sum(1 for ch in w if ch in "bcdfghjklmnpqrstvwxyz")

df = pd.DataFrame({
    "word": words,
    "vowels": [count_vowels(w) for w in words],
    "consonants": [count_consonants(w) for w in words]
})

pivot = pd.crosstab(df["vowels"], df["consonants"])

#1. Heatmap
plt.figure(figsize=(10,5))
sns.heatmap(pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Heatmap Kombinasi Jumlah Vokal–Konsonan per Kata")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()
```

##Visualisasi Headmap
input yang memanggil "*text*" akan mengekstrak semua kata dengan mengambil rangkaian huruf dari A sampai Z menghasilkan heatmap untuk menganalisis pola huruf dalam teks, khususnya jumlah vokal dan konsonan di setiap kata. Output yang di hasilkan memperlihatkan distribusi kombinasi jumlah vokal dan konsonan di seluruh kata dengan nilai numerik di setiap kotak. Warna biru tua menjadi gambaran kombinasi yang paling sering muncul, sedangkan warna kuning menandakan yang jarang muncul. tampak bahwa kata dengan 1 vokal dan 2 konsonan paling sering muncul (sebanyak 71 kata), disusul kombinasi 1 vokal dan 1 konsonan, serta 3 vokal dan 5 konsonan. konsonan-vokal yang kuantitatif yang bersifat pendek.saya jabarkan hasil lebih lanjut outputnya akan menghasilkan sebagai berikut.

1. Kombinasi dengan frekuensi tertinggi (warna biru tua) adalah:

- 1 vokal dan 2 konsonan (71 kata)

- 1 vokal dan 1 konsonan (65 kata)

2. Kombinasi lain yang cukup sering muncul:

- 2 vokal dan 4 konsonan (36 kata)

- 3 vokal dan 5 konsonan (53 kata)

```python
# 2. 3D Surface Plot
X, Y = np.meshgrid(pivot.columns, pivot.index)
Z = pivot.values

fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap="plasma", edgecolor='k', alpha=0.9)
ax.set_xlabel("Jumlah Konsonan")
ax.set_ylabel("Jumlah Vokal")
ax.set_zlabel("Frekuensi")
ax.set_title("Distribusi Frekuensi Kata (3D Surface)")
fig.colorbar(surf, shrink=0.6, aspect=10, pad=0.1)
plt.show()
```

##Surface plot 3D
Pada distribusi Frekuensi Kata 3D ini menunjukan puncak tertinggi berwana (kuning) jenis kata dengan struktur tersebut Sumbu X menunjukkan jumlah huruf konsonan dalam setiap kata, sumbu Y menunjukkan jumlah huruf vokal, dan sumbu Z (tinggi permukaan) menunjukkan frekuensi kemunculan kata dengan kombinasi tersebut. Permukaan grafik berwarna dari ungu hingga kuning menggunakan skema warna plasma di mana warna kuning menandakan frekuensi tertinggi, sedangkan ungu menunjukkan frekuensi terendah.

Grafik 3D memperlihatkan permukaan dengan puncak (warna kuning) di area:

- 1 vokal dan 2 konsonan, frekuensi tertinggi 71 kata.

- 1 vokal dan 1 konsonan, 65 kata.

Bagian bawah (warna ungu gelap) menunjukkan kombinasi yang jarang terjadi, misalnya kata dengan lebih dari 6 konsonan atau lebih dari 4 vokal.

###3. Jumlah kata dalam kalimat

```python
sentences = re.split(r'[.!?]+', text)
sentence_lengths = [len(re.findall(r"[a-zA-Z]+", s)) for s in sentences if s.strip()]
df_sent = pd.DataFrame({"sentence_index": range(1, len(sentence_lengths)+1),
                        "sentence_length": sentence_lengths})
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df_sent["sentence_length"], color="lightblue", width=0.3)
sns.swarmplot(y=df_sent["sentence_length"], color="darkblue", alpha=0.7)
plt.title("Boxplot + Swarmplot: Distribusi Jumlah Kata per Kalimat")
plt.ylabel("Jumlah Kata per Kalimat")
plt.show()
```

##Boxplot + Swarmplot
Gambar diatas merupakan gabungan dari boxplot dan swarmplot. Visuaisasi diatas menampilkan distribusi jumlah kata per kalimat yang bertujuan untuk melihan panjang kalimat yang bervariasi di seluruh teks.

Seperti pada gambar diatas, sebagian besar kalimat memiliki panjang antara sekitar 20 hingga 40 kata dengan median 30 (26-27) kata per kalimat tetapi terdapat juga outlier diatas 60 kata per kalimat yang dimana nilai minimum sekitar 13 kata, sedangkan nilai maksimum mencapai sekitar 65 kata.
Rentang interkuartil (IQR) menunjukkan bahwa 50% kalimat memiliki panjang antara 22 hingga 38 kata.

###Jumlah kalimat dalam paragraf

```python
paragraphs = [p.strip() for p in text.split("\n") if p.strip()]
sentences_per_paragraph = [len(re.split(r'[.!?]+', p)) - 1 for p in paragraphs]
df_para = pd.DataFrame({"paragraph_index": range(1, len(sentences_per_paragraph)+1),
                        "sentence_count": sentences_per_paragraph})

matrix = np.zeros((len(df_para), len(df_para)))
for i in range(len(df_para)):
    for j in range(len(df_para)):
        matrix[i, j] = abs(df_para["sentence_count"][i] - df_para["sentence_count"][j])

plt.figure(figsize=(8,6))
sns.heatmap(matrix, annot=True, fmt=".0f", cmap="YlGnBu",
            xticklabels=df_para["paragraph_index"],
            yticklabels=df_para["paragraph_index"])
plt.title("Heatmap Variasi Jumlah Kalimat Antar Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Paragraf ke-")
plt.show()
```

##Heatmap
Heatmap diatas menampilkan variasi jumlah kalimat antar paragraf dalam sebuah teks. Setiap kotak pada heatmap menunjukkan selisih jumlah antar kalimat antara dua paragraf yang dibandingkan. 0 pada diagonal utama menunjukkan paragraf sedang dibandingkan dengan paragraf itu sendiri sehingga menjadikannya 0 atau tidak valid. Adapun semakin besar nilainya berarti perbedaan jumlah kalimatnya juga semakin besar. Warna heatmap menggambarkan tingkat variasinya.

- Ukuran matriks 13×13 menunjukkan bahwa teks terdiri dari 13 paragraf.

- Nilai selisih antar paragraf berkisar antara 0 hingga 3.

- Warna biru gelap (nilai 3) menunjukkan pasangan paragraf yang paling berbeda jumlah kalimatnya.

- Warna hijau muda hingga kuning (nilai 0–1) menunjukkan paragraf yang panjang kalimatnya relatif sama.

```python
plt.figure(figsize=(9,6))

sizes = df_para["sentence_count"] *
colors = df_para["sentence_count"]

plt.scatter(df_para["paragraph_index"], df_para["sentence_count"],
            s=sizes, c=colors, cmap="coolwarm", alpha=0.7, edgecolors="black")

plt.title("Bubble Chart: Panjang Paragraf vs Posisi Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Jumlah Kalimat")
plt.colorbar(label="Jumlah Kalimat")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

##Bubble Chart
Gambar diatas merupakan buble chart yang menggambarkan hubungan antara posisi paragraf dan jumlah kalimat pada paragraf. Gelembung pada grafik mewakili paragraf dalam teks.

Y menunjukkan jumlah kalimat yang terdapat dalam paragraf dengan jumlah kalimat dalam setiap paragraf (0–3) dan X menunjukkan urutan paragraf (1–13). Dari chart diatas dapat diketahui bahwa paragraf pertama dan pargraf 12 memiliki jumlah kalimat terbanyak yang ditandai dengan gelembung besar berwarna merah. Beberapa paragraf lain seperti 2 dan 9 hanya memiliki satu kalimat yang dapat dilihat dari gelembung kecil berwarna biru.
Ukuran dan warna bubble menunjukkan panjang paragraf (semakin besar dan merah maka semakin banyak kalimat).
