```python
import PyPDF2
import io

pdf_file_path = '/content/Artikel_Matdis_.pdf'

def extract_text_from_pdf(pdf_path):
    """Ekstrak teks dari semua halaman PDF."""
    text = ""
    try:
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text()
    except FileNotFoundError:
        print(f'Error: File tidak ditemukan di {pdf_file_path}')
        return None
    except Exception as e:
        print(f'Terjadi kesalahan saat membaca PDF: {e}')
        return None
    return text

def find_section_text(text, start_keywords, end_keywords):
    """Temukan teks di antara kata kunci awal dan akhir."""
    start_index = -1
    for keyword in start_keywords:
        start_index = text.lower().find(keyword.lower())
        if start_index != -1:
            start_index += len(keyword)
            break

    if start_index == -1:
        return None

    end_index = -1
    for keyword in end_keywords:
        end_index = text.lower().find(keyword.lower(), start_index)
        if end_index != -1:
            break

    if end_index == -1:
        return text[start_index:]
    else:
        return text[start_index:end_index]

full_text = extract_text_from_pdf(pdf_file_path)

if full_text:
    abstract_start_keywords = ["abstract", "abstrak"]
    introduction_start_keywords = ["introduction", "pendahuluan"]
    conclusion_start_keywords = ["conclusion", "kesimpulan"]

    abstract_end_keywords = ["introduction", "pendahuluan", "keywords", "kata kunci"]
    introduction_end_keywords = ["methodology", "metodologi", "metode", "related work", "penelitian terkait", "result", "hasil"]
    conclusion_end_keywords = ["references", "daftar pustaka", "acknowledgments", "ucapan terima kasih"]

    abstract_text = find_section_text(full_text, abstract_start_keywords, abstract_end_keywords)
    introduction_text = find_section_text(full_text, introduction_start_keywords, introduction_end_keywords)
    conclusion_text = find_section_text(full_text, conclusion_start_keywords, conclusion_end_keywords)

    if abstract_text:
        print('--- Panjang Teks Abstrak ---')
        print(f'{len(abstract_text.strip())} karakter')
        print('\n' + '='*20 + '\n')

    else:
        print('Bagian Abstrak tidak ditemukan.')

    if introduction_text:
        print('--- Panjang Teks Pendahuluan ---')
        print(f'{len(introduction_text.strip())} karakter')
        print('\n' + '='*20 + '\n')
    else:
        print('Bagian Pendahuluan tidak ditemukan.')

    if conclusion_text:
        print('--- Panjang Teks Kesimpulan ---')
        print(f'{len(conclusion_text.strip())} karakter')
        print('\n' + '='*20 + '\n')
    else:
        print('Bagian Kesimpulan tidak ditemukan.')
```

```python
import re

def clean_text(text):
    """Membersihkan teks dari tanda baca, angka, dan simbol."""
    if text is None:
        return None
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

cleaned_abstract_text = clean_text(abstract_text)
cleaned_introduction_text = clean_text(introduction_text)
cleaned_conclusion_text = clean_text(conclusion_text)

if cleaned_abstract_text:
    print('--- Panjang Teks Abstrak (Setelah Pembersihan) ---')
    print(f'{len(cleaned_abstract_text)} karakter')
    print('\n' + '='*20 + '\n')
else:
    print('Abstrak kosong setelah pembersihan.')

if cleaned_introduction_text:
    print('--- Panjang Teks Pendahuluan (Setelah Pembersihan) ---')
    print(f'{len(cleaned_introduction_text)} karakter')
    print('\n' + '='*20 + '\n')
else:
    print('Pendahuluan kosong setelah pembersihan.')

if cleaned_conclusion_text:
    print('--- Panjang Teks Kesimpulan (Setelah Pembersihan) ---')
    print(f'{len(cleaned_conclusion_text)} karakter')
    print('\n' + '='*20 + '\n')
else:
    print('Kesimpulan kosong setelah pembersihan.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

text_to_analyze = cleaned_introduction_text

if text_to_analyze:
    words = re.findall(r"[a-zA-Z]+", text_to_analyze.lower())

    def count_vowels(w): return sum(1 for ch in w if ch in "aeiou")
    def count_consonants(w): return sum(1 for ch in w if ch in "bcdfghjklmnpqrstvwxyz")

    df_words = pd.DataFrame({
        "word": words,
        "vowels": [count_vowels(w) for w in words],
        "consonants": [count_consonants(w) for w in words]
    })

    pivot_frequency = pd.crosstab(df_words["vowels"], df_words["consonants"])

    pivot_count = pivot_frequency

    # 1. Heatmap Frekuensi
    plt.figure(figsize=(12, 6))
    sns.heatmap(pivot_frequency, cmap="YlGnBu", annot=True, fmt="d")
    plt.title("Heatmap Frekuensi Kombinasi Jumlah Vokal–Konsonan per Kata")
    plt.xlabel("Jumlah Konsonan")
    plt.ylabel("Jumlah Vokal")
    plt.show()

    print("\n--- Tabel Pivot Frekuensi Kombinasi Vokal-Konsonan ---")
    display(pivot_frequency)

else:
    print("Teks untuk analisis kosong.")
```

###3. Jumlah kalimat dalam paragraf

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PyPDF2 import PdfReader

pdf_path = "Artikel_Matdis_.pdf"
reader = PdfReader(pdf_path)
full_text = " ".join([page.extract_text() for page in reader.pages if page.extract_text()])

pattern = re.compile(
    r'(Abstract.*?)(?=Introduction|1\.|I\.)'
    r'|(Introduction.*?)(?=Conclusion|References|Bibliography|Acknowledg)'
    r'|(Conclusion.*?)(?=References|Bibliography|Acknowledg)',
    re.IGNORECASE | re.DOTALL
)
sections = pattern.findall(full_text)
selected_text = " ".join([" ".join(s) for s in sections if any(s)]).strip()

if not selected_text:
    print("⚠️ Tidak ditemukan bagian Abstract, Introduction, atau Conclusion di artikel ini.")
else:
    cleaned_text = re.sub(r'-\s*\n', '', selected_text)
    cleaned_text = re.sub(r'\n+', ' ', cleaned_text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    sentence_pattern = re.compile(r'(?<=[.!?])\s+(?=[A-Z(“"])')
    sentences = re.split(sentence_pattern, cleaned_text)

    paragraphs = []
    current = []
    for s in sentences:
        current.append(s)
        if len(" ".join(current).split()) > 60:
            paragraphs.append(" ".join(current))
            current = []
    if current:
        paragraphs.append(" ".join(current))

    para_sentence_counts = [len(re.split(sentence_pattern, p)) for p in paragraphs]

    df_para = pd.DataFrame({
        "paragraph_index": range(1, len(para_sentence_counts) + 1),
        "sentence_count": para_sentence_counts,
        "paragraph_text": paragraphs
    })

    longest = df_para.loc[df_para["sentence_count"].idxmax()]

    print("📘 Paragraf dengan jumlah kalimat terbanyak (Abstract, Introduction, Conclusion):")
    print(f"➡️ Paragraf ke-{longest['paragraph_index']} dengan {longest['sentence_count']} kalimat.\n")
    print("Isi Paragraf (potongan):\n")
    print(longest['paragraph_text'][:700] + "...\n")

    plt.figure(figsize=(8, 5))
    sns.histplot(df_para["sentence_count"], bins=10, kde=True, color="lightblue", edgecolor="black")
    plt.title("Distribusi Jumlah Kalimat per Paragraf (Abstract, Introduction, Conclusion)")
    plt.xlabel("Jumlah Kalimat per Paragraf")
    plt.ylabel("Frekuensi")
    plt.show()
```

Sumbu-X pada grafik menggambarkan jumlah kalimat yang terdapat dalam setiap paragraf, mulai dari satu hingga lima kalimat. Sementara itu, sumbu-Y menunjukkan frekuensi atau seberapa banyak paragraf yang memiliki jumlah kalimat tertentu.
